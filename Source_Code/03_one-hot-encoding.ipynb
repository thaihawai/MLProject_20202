```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/data_shortened.csv', encoding='utf-8')
df.fillna('', inplace=True)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df.replace({'employment_type': {'': 'Other'}, 'required_experience': {'': 'Other', 'Not Applicable': 'Other'}}, inplace=True)
```

```python
# one hot encoding these two columns
to_onehot = ['employment_type', 'required_experience']
for item in to_onehot:
    ohe = OneHotEncoder()
    ohe.fit(df[item].to_numpy().reshape(-1,1))
    transformed = ohe.transform(df[item].to_numpy().reshape(-1,1))
    ohe_df = pd.DataFrame.sparse.from_spmatrix(transformed)
    ohe_df.columns = ohe.get_feature_names([item])
    df = pd.concat([df, ohe_df], axis=1).drop([item], axis=1)
```

```python
# the first 6 columns are text data
df.iloc[:, 6:].astype(int).to_csv('Data/data_all_categorical.csv', encoding='utf-8', index=False)
```
